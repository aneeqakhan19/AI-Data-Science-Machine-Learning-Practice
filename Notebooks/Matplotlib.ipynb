```python
x = [1,2,3,4,5,6,7]
y = [6,9,45,64,50,61,70]
plt.plot(x,y)
plt.show()
```

## Different Methods to Plot

### Method-1

```python
home = plt.figure()
room = home.add_subplot()
room.plot(x,y)
plt.show()
```

### Method-2

```python
home = plt.figure()
room = home.add_axes([1,1,1,1])
room.plot(x,y)
plt.show()
```

### Method-3  (More Robust and Compact) 

```python
home , room = plt.subplots()
room.plot(x,y)
plt.show()
```

## Setting-up features of plot 

```python
fig , sub_plot = plt.subplots()
sub_plot.plot(x,y)
sub_plot.set(title='A GOOD PLOT',xlabel = 'x-axis', ylabel = 'y-axis')
plt.show()
```

## pyplot Vs Object Oriented Method 

```python
heart = pd.read_csv('Heart disease data.csv')
```

```python
heart
```

```python
heart.head()
```

```python
over_40 = heart[heart['age']>40]
```

```python
over_40
```

### pyplot

```python
over_40.plot(kind='scatter', x='age',y='chol',c='condition');
```

### object-oriented approach

```python
import matplotlib.pyplot as plt
fig , ax = plt.subplots(figsize = (6,5))
over_40.plot(kind='scatter',x='age',y='chol',c='condition',ax=ax);
```

## Entire Life Cycle of Object Oriented Plotting

```python
#create fig and graph area
fig , ax = plt.subplots(figsize = (7,7))
#plot the data 
suspect = ax.scatter(x=over_40['age'],y=over_40['chol'],c=over_40['condition']);
#customization
ax.set(title='Suspectibilty of Heart Disease',xlabel = 'Age', ylabel = 'cholestrol');
#legend
ax.legend(*suspect.legend_elements(),title = 'CONDITION');
suspect
```

## Entire Life Cycle of Object Oriented Plotting (ADVANCED)

```python
plt.style.use('dark_background')
#create plot area 
fig , (plot1,plot2) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
#add data to plot 1 
suspect1 = plot1.scatter(x=over_40['age'],y=over_40['chol'],c=over_40['condition'],cmap='winter');#to change color of data
#customization
plot1.set(title='Heart Disease Problem depending on Age and Cholestrol', ylabel = 'CHOLESTROL');
plot1.legend(*suspect1.legend_elements(),title = 'Survived or Not');
suspect1
#add data to plot 2
suspect2 = plot2.scatter(x=over_40['age'],y=over_40['trestbps'],c=over_40['condition'],cmap='plasma');
#customization
plot2.set(title='Heart Disease Problem depending on Age and trestbps',xlabel = 'AGE', ylabel = 'trestbps',label= over_40['condition'])
plot2.legend(*suspect2.legend_elements(),title = 'Survived or Not');
suspect2
#meanline 
plot1.axhline(y=over_40['chol'].mean(),linestyle='--',color='white')
plot2.axhline(y=over_40['trestbps'].mean(),linestyle='--',color='red')
#figure styling
plt.style.use('dark_background')
#naming entire figure 
fig.suptitle("PATIENTS DATA",fontsize=32,fontweight='bold')
#save figure 
fig.savefig('Patients Data Report.png')
plt.show()
```
